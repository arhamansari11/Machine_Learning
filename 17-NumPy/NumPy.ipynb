```python
arr = np.array([1 , 2 ,3 ])
arr
```

```python
arr1 = np.array=([[1 , 2 ,3] , [4 ,5 ,6]])
arr1
```

```python
zero_arr = np.zeros(8)
one_arr = np.ones(4)
print(zero_arr)
one_arr
```

```python
range_arr =  np.arange(8)
range_arr
```

```python
reshap1 = np.arange(9)
reshap1 = reshap1.reshape(3 , 3)
reshap1
```

```python
sample_arr = [8 , 9 ,10 , 11 , 12 , 13 , 14]
sample_arr2 = [ 1 , 2, 3 , 4 ,5 ,6 ,7]
print(sample_arr , sample_arr2)

plus = np.multiply(sample_arr , sample_arr2)
print(plus)
```

```python
stat = [8 , 9 ,10 , 11 , 12 , 13 , 14]
print(np.sum(stat))
print(np.min(stat))
print(np.max(stat))
print(np.mean(stat))
```

```python
indexing = [8 , 9 ,10 , 11 , 12 , 13 , 14]
print(indexing[0:3])
```

```python
random = np.random.rand(4) + 2
for i in random:
    print(i)
```
